```python
# This piece of logic tilts the image 180 degress . With this code base we can augment the data by tilting every image. 
# Since this is poster images only , this might not be pretty useful but for anyother image classification we can use this

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


im = np.array(Image.open('poster_0.jpeg'),'f')
im_orig = Image.fromarray(im.astype('uint8'))
#print im.shape

# Rotate the image 180 degrees like a mirror image
im1 = np.copy(im)
for d in range(im.shape[2]):
    for h in range(im.shape[0]):
        for w in range(im.shape[1]):
            im1[h][w][d] = im[h][im.shape[1] -1 -w][d]

im2 = Image.fromarray(im1.astype('uint8'))

#PLot  to show an image before and after

fig = plt.figure()
a=fig.add_subplot(1,2,1)
a.set_title('Before')
plt.imshow(im_orig)
a=fig.add_subplot(1,2,2)
a.set_title('After')
plt.imshow(im2)
plt.show()
```
